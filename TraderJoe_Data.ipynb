```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

states = []

url = "https://locations.traderjoes.com/"
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

div_tags= soup.find_all("div", class_="itemlist")

for div_tag in div_tags:
  a_tag = div_tag.find("a")
  href = a_tag["href"]
  state = a_tag.get_text()

  temp = href.split("/")[-2]
  states.append(temp)
  #print(temp)
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

city_urls = []

for state in states:
  url = f"https://locations.traderjoes.com/{state}/"
  response = requests.get(url)
  soup = BeautifulSoup(response.content, 'html.parser')

  div_tags = soup.find_all("div",  class_="itemlist")

  for div_tag in div_tags:
    a_tag = div_tag.find("a")
    href = a_tag["href"]
    county = a_tag.get_text()
    #print(state, county, ":", href)

    city_urls.append(href)
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

branch_urls = []

for city_url in city_urls:
  response = requests.get(city_url)
  soup = BeautifulSoup(response.content, 'html.parser')

  div_tags = soup.find_all("div", class_="address-left")

  for div_tag in div_tags:
    a_tag = div_tag.find("a")
    url = a_tag["href"]
    branch_urls.append(url)
    print(url)
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re

df_url = []
df_address = []
df_city = []
df_state = []
df_zip = []
df_phone = []

for branch_url in branch_urls:
    response = requests.get(branch_url)
    soup = BeautifulSoup(response.content, 'html.parser')

    div_tag = soup.find("div", class_="addressline")
    address = div_tag.get_text().strip()
    address= address.replace('\t\t\t\t\t\t\t         \t', "")
    address= address.replace('\t            ', "")
    address= address.replace('     \t ', "")
    address= address.replace(', ', "")

    address = address.strip()

    temp = address.split("\n")

    line_1 = temp[0]
    city = temp[2]
    state = temp[3]
    zip = temp[4]
    phone = temp[-1]

    df_url.append(branch_url)
    df_address.append(line_1)
    df_city.append(city)
    df_state.append(state)
    df_zip.append(zip)
    df_phone.append(phone)

    #print(line_1, ":", city, state, zip, phone)
```

```python
df = pd.DataFrame({"State":df_state,
                   "City":df_city,
                   "Address":df_address,
                   "Zip":df_zip,
                   "URL":df_url,
                   "Phone":df_phone})
df.to_csv("trader_joe_branches.csv", index=False)
```

```python
df_count = df.groupby("State")["Address"].count().reset_index()

df_count.columns = ["State", "Count"]

df_count
```
